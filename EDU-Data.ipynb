```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('Edu-Data.csv')
print(df.head())
print(df.columns)
print(df.info())


#1. How many male and female students are there?
Gender = df['gender'].value_counts()
print("---there are this many male and female---")
print(Gender)

#What is the average number of “raisedhands” for each class (M, L, H)?
raisedhand_class = df.groupby('Class')["raisedhands"].mean()
print("---the average number of raisedhands in each class---")
print(raisedhand_class)

#Which Nationality (NationalITy) group has the most students?
nati_group = df['NationalITy'].value_counts()
print("---Nationality groups---")
print(nati_group)

#What is the average number of VisITedResources for each gender?
visited_gender = df.groupby('gender')['VisITedResources'].mean()
print("---visitedresources for each gender---")
print(visited_gender)

#Which Topic had the highest average Discussion score?
topic_score = df.groupby('Topic')['Discussion'].mean()
print("---Highest dicussion score for each topic---")
print(topic_score)

#How many students have “Good” ParentSchoolSatisfaction?
student_good = (df['ParentschoolSatisfaction'] == 'Good').sum()
print("--- students that meet parents satisfaction---")
print(student_good)

#Which “StageID” (school level) has the highest average “raisedhands”?
stage_averg = df.groupby('StageID')['raisedhands'].mean()
print("---average raisedhands of StageID----")
print(stage_averg)

#What is the most common “Class” (H, M, L) among students whose ParentSchoolSatisfaction is “Bad”?
most_common = df[df['ParentschoolSatisfaction'] == 'Bad']['Class'].value_counts()
print("---student who have bad parent satisfaction---")
print(most_common)

#Which “SectionID” (A, B, or C) had the highest average math “GradeID”
highest_averg = df.groupby('SectionID')['raisedhands'].mean().sort_values(ascending=False)
print("---highest average math grade ID section ID---")
print(highest_averg)

#What is the average number of AnnouncementsView for each Relation (Father, Mother, etc.)?
each_relation = df.groupby('Relation')['AnnouncementsView'].mean()
print("--- AnnouncementsView for each relation---")
print(each_relation)

#How many students have “Above-7” StudentAbsenceDays and are in “Class” H?
count = len(df[(df['StudentAbsenceDays'] == 'Above-7') & (df['Class'] == 'H')])
print("---", count, "----")

#For each Topic, what is the maximum value of raisedhands?
each_topic = df.groupby('Topic')['raisedhands'].idxmax()
print("---maxi---", each_topic)

df['gender'].value_counts().plot(kind='bar')
plt.show()

df['Class'].value_counts().plot(kind='bar')
plt.show()
df.to_csv('my_resultd.csv', index=False)
```
